```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
```

```python
df = pd.read_csv("data\output\problem2_cluster_summary.csv")
```

```python
df.head(10)
```

```python
df["cluster_id"] = df["cluster_id"].astype(str)
```

```python
#set our colors
cmap = plt.get_cmap("tab10")  
colors = [cmap(i % cmap.N) for i in range(len(df))]

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(
    df["cluster_id"],
    df["num_applications"],
    color = colors
)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f"{int(height)}",
        ha="center",
        va="bottom"
    )

# Labels and title
plt.xlabel("Cluster ID")
plt.ylabel("Number of Applications")
plt.title("Number of Applications per Cluster")

plt.tight_layout()

# Save figure
plt.savefig("problem2_bar_chart.png")
plt.close()
```

```python
#now plot density... 
df2 = pd.read_csv("data\output\problem2_timeline.csv")
```

```python
#find largest cluster ID..
largest_cluster_id = (df2.groupby("cluster_id").size().idxmax())

cluster_df = df2[df2["cluster_id"] == largest_cluster_id]
```

```python
durations = cluster_df["duration_sec"].dropna()
durations = durations[durations > 0]
n = len(durations)
```

```python
#plot our histogram with KDE overlay...
plt.figure(figsize=(10, 6))
plt.hist(
    durations,
    bins=30,
    density=True,
    alpha=0.6
)

#kde portion...
kde = gaussian_kde(durations)
x_vals = np.logspace(
    np.log10(durations.min()),
    np.log10(durations.max()),
    500
)
plt.plot(x_vals, kde(x_vals))

#Set our log scale on x-axis
plt.xscale("log")

#set labels for axis 
plt.xlabel("Job Duration (seconds, log scale)")
plt.ylabel("Density")
plt.title(
    f"Job Duration Distribution for Largest Cluster "
    f"(Cluster {largest_cluster_id}, n={n})"
)

plt.tight_layout()
plt.savefig("problem2_density_plot.png")
plt.close()
```
